```python
"""
importing the necessaCreditScoring.csv"""
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('CreditScoring.csv') 
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
status_values = {
    1: 'ok',
    2: 'default',
    0: 'unk'
}

df.status = df.status.map(status_values)


home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}

df.home = df.home.map(home_values)

marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}

df.marital = df.marital.map(marital_values)

records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}

df.records = df.records.map(records_values)

job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}

df.job = df.job.map(job_values)
```

```python
df.head()
```

```python
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace=99999999, value=0)
```

```python
df = df[df.status != 'unk'].reset_index(drop=True)
```

```python
df['default'] = (df.status == 'default').astype(int)
del df['status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
print("income:", round(roc_auc_score(df_train['default'], df_train['income']),2))
print("assets:", round(roc_auc_score(df_train['default'], df_train['assets']),2))
print("debt:", round(roc_auc_score(df_train['default'], df_train['debt']),2))
print("seniority:", round(roc_auc_score(df_train['default'], df_train['seniority']),2))
```
